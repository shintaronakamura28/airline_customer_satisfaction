## EDA 

```python
def explore_data(df):
    number_rows = df.shape[0]
    number_features = df.shape[1]
    data_types = df.dtypes

    print(f'Number of rows: {number_rows}')
    print(f'\nNumber of columns: {number_features}')
    print(f'\nData types: ')
    print(data_types)

explore_data(df)
```

```python
#functions to perform EDA
def perform_eda(df):
    duplicates = df.duplicated().sum()
    missing_values = df.isna().sum()

    print(f'Number of Duplicate Rows: {duplicates}')
    print(f'\nNumber of Missing Values: {missing_values}')

perform_eda(df)
```

```python
#function to check all unique values and anomalies

def check_col_values(df):
    print('Check all columns for anomalies \n')

    #loop throguh each column i nthe data frame
    for column_name in df.columns:
        print(f'Unique Values in {column_name}')
        print(df[column_name].unique())
        print('\n')

check_col_values(df)
```
