## EDA 

```python
def explore_data(df):
    number_rows = df.shape[0]
    number_features = df.shape[1]
    data_types = df.dtypes

    print(f'Number of rows: {number_rows}')
    print(f'\nNumber of columns: {number_features}')
    print(f'\nData types: ')
    print(data_types)

explore_data(df)
```

```python
#functions to perform EDA
def perform_eda(df):
    duplicates = df.duplicated().sum()
    missing_values = df.isna().sum()

    print(f'Number of Duplicate Rows: {duplicates}')
    print(f'\nNumber of Missing Values: {missing_values}')

perform_eda(df)
```

## Visualization 

```python
def plot_correlation_heatmap(df):
    fig, ax = plt.subplots(figsize= (15,5))

    #define the mask to set the values in the upper traingle to false
    mask = np.triu(np.ones_like(df.corr(), dtype=np.bool))

    #plot the heatmap
    sns.heatmap(
        df.corr().round(2),
        ax=ax,
        mask=mask,
        vmin=-1,
        vmax=1,
        annot=True,
        cmap='BrBG'
    )

    ax.set_title('Triangle Correlation Heatmap', fontdict={'fontsize': 20}, pad=16)
    plt.show()

plot_correlation_heatmap(df)
```

```python
# Correlation heatmap as ranked
def ranked_correlation_heatmap(df, target_column):
    plt.figure(figsize= (8,8))
    heatmap = sns.heatmap(
        df.corr()[[target_column]].sort_values(by=target_column, ascending=False),
        vmin= -1, 
        vmax = 1, 
        annot = True, 
        cmap= 'BrBG'
    )
    
    heatmap.set_title(f'Features Correlation with {target_column}', fontdict={'fontsize': 12}, pad=10)
    plt.show()

ranked_correlation_heatmap(df, target_column='satisfied')
```

```python
#Create histogram and boxplot
def visualize_data(df, column_name):
    fig, axes = plt.subplots(nrows=2, sharex=False, figsize = (6,6))
    sns.set_style('dark')

    #histogram plot
    sns.histplot(df[column_name], ax=axes[0], bins='auto', kde=True)
    axes[0].tick_params(axis='x', labelrotation= 45)


    #box plot
    sns.boxplot(data=df, x=column_name, ax=axes[1])
    plt.tight_layout()
    plt.show

visualize_data(df, column_name='Flight Distance')
```
